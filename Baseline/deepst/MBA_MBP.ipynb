```python
import os
from DeepST import run
import matplotlib.pyplot as plt
from pathlib import Path
import scanpy as sc
import numpy as np
from sklearn import metrics
import json
import time
import os.path as osp
import pandas as pd

os.environ["CUDA_VISIBLE_DEVICES"] = "0"
os.environ['R_HOME'] = r'C:\Program Files\R\R-4.4.2'
os.environ['R_USER'] = r'C:\Users\CLEARLOVE\.conda\envs\STG\Lib\site-packages\rpy2'
```

```python
data_path = r"C:\Code\Data\ST Adult mouse brain section A\MBA-MBP"
data_name_list = ['Mouse_Brain_Anterior', 'Mouse_Brain_Posterior']
save_path = "C:\Code\STG3Net-main\Baseline\deepst_result"
n_domains = 32

deepen = run(save_path = save_path,
	task = "Integration",
	pre_epochs = 400,
	epochs = 500,
	use_gpu = True,
	)

###### Generate an augmented list of multiple datasets
augement_data_list = []
graph_list = []
for i in range(len(data_name_list)):
	adata = deepen._get_adata(platform="Visium", data_path=data_path, data_name=data_name_list[i])
	adata = deepen._get_image_crop(adata, data_name=data_name_list[i])
	adata = deepen._get_augment(adata, spatial_type="LinearRegress")
	graph_dict = deepen._get_graph(adata.obsm["spatial"], distType = "KDTree", k=12)
	augement_data_list.append(adata)
	graph_list.append(graph_dict)

######## Synthetic Datasets and Graphs
multiple_adata, multiple_graph = deepen._get_multiple_adata(adata_list = augement_data_list, data_name_list = data_name_list, graph_list = graph_list)

###### Enhanced data preprocessing
data = deepen._data_process(multiple_adata, pca_n_comps = 200)

deepst_embed = deepen._fit(
		data = data,
		graph_dict = multiple_graph,
		domains = multiple_adata.obs["batch"].values,  ##### Input to Domain Adversarial Model
		n_domains = len(data_name_list))
multiple_adata.obsm["DeepST_embed"] = deepst_embed
multiple_adata = deepen._get_cluster_data(multiple_adata, n_domains=n_domains, priori = True)

sc.pp.neighbors(multiple_adata, use_rep='DeepST_embed')
sc.tl.umap(multiple_adata)
sc.pl.umap(multiple_adata, color=["DeepST_refine_domain","batch_name"])
# plt.savefig(os.path.join(save_path, f'{"_".join(data_name_list)}_umap.pdf'), bbox_inches='tight', dpi=300)

for data_name in data_name_list:
	adata = multiple_adata[multiple_adata.obs["batch_name"]==data_name]
	sc.pl.spatial(adata, color='DeepST_refine_domain', frameon = False, spot_size=150)
# 	plt.savefig(os.path.join(save_path, f'{data_name}_domains.pdf'), bbox_inches='tight', dpi=300)
```

```python
import matplotlib.pyplot as plt

plt.rcParams['axes.facecolor'] = 'white'  # 设置子图的背景颜色
plt.rcParams['figure.facecolor'] = 'white'  # 设置整张图的背景颜色

adata_anterior = multiple_adata[multiple_adata.obs['batch_name'] == 'Mouse_Brain_Anterior'].copy()
adata_posterior = multiple_adata[multiple_adata.obs['batch_name'] == 'Mouse_Brain_Posterior'].copy()

# Anterior 部分
spatial_a = pd.read_csv(
    r"C:\Code\Data\ST Adult mouse brain section A\MBA-MBP/Mouse_Brain_Anterior/spatial/tissue_positions_list.csv",
    sep=",", header=None, na_filter=False, index_col=0
)
adata_anterior.obs["x_array"] = spatial_a[2]
adata_anterior.obs["y_array"] = spatial_a[3]
adata_anterior.obs["x_pixel"] = spatial_a[4]
adata_anterior.obs["y_pixel"] = spatial_a[5]

# Posterior 部分
spatial_p = pd.read_csv(
    r"C:\Code\Data\ST Adult mouse brain section A\MBA-MBP/Mouse_Brain_Posterior/spatial/tissue_positions_list.csv",
    sep=",", header=None, na_filter=False, index_col=0
)
adata_posterior.obs["x_array"] = spatial_p[2]
adata_posterior.obs["y_array"] = spatial_p[3]
adata_posterior.obs["x_pixel"] = spatial_p[4]
adata_posterior.obs["y_pixel"] = spatial_p[5]

# 计算对齐偏移量（假设 anterior 不动，posterior 向右平移）
x_offset = adata_anterior.obs["x_pixel"].max() - adata_posterior.obs["x_pixel"].min() + 100  # 100 是间距
y_offset = adata_anterior.obs["y_pixel"].max() - adata_posterior.obs["y_pixel"].min() + 100

# 更新 posterior 坐标
adata_posterior.obs["x_pixel"] += x_offset
adata_posterior.obs["y_pixel"] += y_offset

# 存储对齐后的坐标到 obsm（可选）
adata_anterior.obsm["spatial_aligned"] = adata_anterior.obs[["x_pixel", "y_pixel"]].values
adata_posterior.obsm["spatial_aligned"] = adata_posterior.obs[["x_pixel", "y_pixel"]].values

# 更新 anterior 部分
anterior_mask = multiple_adata.obs['batch_name'] == 'Mouse_Brain_Anterior'
multiple_adata.obs.loc[anterior_mask, "x_pixel"] = adata_anterior.obs["x_pixel"]
multiple_adata.obs.loc[anterior_mask, "y_pixel"] = adata_anterior.obs["y_pixel"]

# 更新 posterior 部分
posterior_mask = multiple_adata.obs['batch_name'] == 'Mouse_Brain_Posterior'
multiple_adata.obs.loc[posterior_mask, "x_pixel"] = adata_posterior.obs["x_pixel"]
multiple_adata.obs.loc[posterior_mask, "y_pixel"] = adata_posterior.obs["y_pixel"]

# 更新 obsm（可选）
multiple_adata.obsm["spatial_aligned"] = multiple_adata.obs[["x_pixel", "y_pixel"]].values

subadata = multiple_adata[~multiple_adata.obs['DeepST_refine_domain'].isna()]  # 删除 NaN
ax=sc.pl.scatter(subadata,alpha=1,x="y_pixel",y="x_pixel",color="DeepST_refine_domain",show=False,size=200000/adata.shape[0])
ax.set_aspect('equal', 'box')
ax.axes.invert_yaxis()
plt.grid(False)
```
